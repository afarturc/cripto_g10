## Observação sobre a Uniformidade e a Distância Total de Variação (TV)

Embora as distribuições obtidas nos casos analisados sejam quase uniformes, **nenhuma delas é perfeitamente uniforme** quando $2^n$ não é múltiplo de $p$. A diferença pode ser quantificada pela **distância total de variação (TV)**, que mede o quão distante uma distribuição está da uniforme:

$$
TV(P, U) = \frac{1}{2} \sum_{s \in S} |P(s) - U(s)|
$$

onde $P(s)$ é a probabilidade real e $U(s)$ é a probabilidade da distribuição uniforme. Quanto menor o valor de TV, mais próxima a distribuição está da uniforme.

````python
# Probabilidades para n = 8, p = 251
p = 251
n = 8
C = 2**n
prob = []
for s in range(p):
    if s < C - p:
        prob.append(2/C if s < C - p else 1/C)
    else:
        prob.append(1/C)
# Ajuste correto:
prob = [2/C if s < C - p else 1/C for s in range(p)]

# prob
for s in range(p):
    print(f"P({s}) = {prob[s]}")
````

````python
# Probabilidades para n = 64, p = 251
p = 251
n = 64
C = 2**n
base = C // p
resto = C % p
prob = [base / C for _ in range(p)]
for s in range(resto):
    prob[s] += 1 / C

# Mostra as probabilidades
for s in range(p):
    print(f"P({s}) = {prob[s]}")
````

````python
# Probabilidades para p = 256 (potência de 2)
# Para n = 8 e n = 64, a distribuição é uniforme
p = 256
n = 8
C = 2**n
prob = [1/p for _ in range(p)]
print('Distribuição uniforme para p = 256, n = 8:')
for s in range(p):
    print(f"P({s}) = {prob[s]}")

n = 64
C = 2**n
prob = [1/p for _ in range(p)]
print('Distribuição uniforme para p = 256, n = 64:')
for s in range(p):
    print(f"P({s}) = {prob[s]}")
````

````python
# Cálculo da entropia usando SageMath

def entropy(prob):
    from math import log2
    return -sum(p * log2(p) for p in prob if p > 0)

# Exemplo para n = 8, p = 251
def prob_n8_p251():
    p = 251
    n = 8
    C = 2**n
    return [2/C if s < C - p else 1/C for s in range(p)]

# Exemplo para n = 64, p = 251
def prob_n64_p251():
    p = 251
    n = 64
    C = 2**n
    base = C // p
    resto = C % p
    prob = [base / C for _ in range(p)]
    for s in range(resto):
        prob[s] += 1 / C
    return prob

# Exemplo para p = 256 (potência de 2)
def prob_p256():
    p = 256
    return [1/p for _ in range(p)]

# calculo da entropia
e_n8_p251 = entropy(prob_n8_p251())
e_n64_p251 = entropy(prob_n64_p251())
e_p256 = entropy(prob_p256())

print(f"Entropia para n=8, p=251: {e_n8_p251}")
print(f"Entropia para n=64, p=251: {e_n64_p251}")
print(f"Entropia para p=256: {e_p256}")
````

# Exercício 4

#### Estratégia

1. Para cada k=1,2,… calcular q,r por divisão inteira: q,r=divmod(2k,p)
2. Avaliar Hk pela fórmula acima (usar aritmética de ponto flutuante de dupla precisão ou racionais).
3. Parar quando 
∣Hk−log⁡2p∣<tolerância. Se queremos igualdade exacta, procurar r=0 em vez de comparar entropias.

````python
# Sage code: entropia por contagem e busca do menor k
import math

def entropy_by_counts(p, k):
    C = 2**k
    q, r = divmod(C, p)
    H = 0.0
    if r > 0:
        pr_big = (q+1)/C
        H -= r * pr_big * math.log(pr_big, 2)
    if p - r > 0 and q > 0:
        pr_small = q/C
        H -= (p - r) * pr_small * math.log(pr_small, 2)
    return H, q, r

def find_min_k_for_entropy(p, tol=1e-9, k_max=2000):
    H_unif = math.log(p, 2)
    for k in range(1, k_max+1):
        Hk, q, r = entropy_by_counts(p, k)
        if abs(Hk - H_unif) < tol:
            return k, Hk, abs(Hk - H_unif), q, r
    return None

# Exemplos:
p1 = 251
p2 = 256

# tolerâncias que podemos testar
tols = [1e-6, 1e-9]

for p in (p1, p2):
    print("p =", p, "H_unif =", math.log(p,2))
    for tol in tols:
        res = find_min_k_for_entropy(p, tol=tol, k_max=2000)
        if res is None:
            print("  tol =", tol, ": não encontrado k ≤ 2000")
        else:
            k, Hk, diff, q, r = res
            print("  tol =", tol, ": k =", k, " Hk =", Hk, " diff =", diff, " q =", q, " r =", r)
    # verificar existência de r=0 (igualdade exacta)
    # r=0 <=> p / 2^k; se p é potência de 2, existe k >= log2(p)
    for ktest in range(1,65):
        q, r = divmod(2**ktest, p)
        if r == 0:
            print("  r=0 para k =", ktest, "=> distribuição uniforme exata")
            break
    else:
        print("  nunca r=0 para k ≤ 64")
````
