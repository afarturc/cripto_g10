### Q2: ECC

#### Question - P1

So far, we know that Alice picks the signing key $sk_{A}$ and forms the public verifying key by computing $pk_{A} \leftarrow sk_{A} \cdot G$.

Then, she picks a random value $k$, and computes the signature $\sigma \leftarrow m - k \cdot sk_{A} \cdot G$. Sends these informations to Bob.

Then Bob checks that $m = \sigma + k \cdot pk_{A}$

Let's then substitute $\sigma$ for $m - k \cdot sk_{A} \cdot G$ in the last equation

$m = \sigma + k \cdot pk_{A} \Leftrightarrow m = m - k \cdot sk_{A} \cdot G + k \cdot pk_{A}$

Let's then substiture $pk_{A}$ with $sk_{A} \cdot G$

$m = m - k \cdot sk_{A} \cdot G + k \cdot sk_{A} \cdot G \Leftrightarrow m = m$

This shows that the verification algorithm works accordingly for correctly signed messages

#### Question - P2

Assuming that at this point, we get all the values except for $sk_{A}$. These value are:
$p, G, pk_{A}, m, k$ and $\sigma$

So, we know that $\sigma \leftarrow m - k \cdot sk_{A} \cdot G$ and that 
$pk_{A} \leftarrow sk_{A} \cdot G$

So, by making a substitution in the $\sigma$ equation we get:

$\sigma \leftarrow m - k \cdot pk_{A}$

$m$, $k$, and $pk_{A}$ are the message, the random value k picked by Alice and the public key computed by Alice respectively. All of these values are public and can be got by any attacker.
