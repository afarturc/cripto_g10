```python
import string

key = "NOTVERYDIFCULMPQSWXZABGHJK"

#Converte uma letra maiúscula (A–Z) para um número de 0 a 25.
def encodeOne(c):
    assert c in string.ascii_uppercase
    return ord(c)-ord('A')

#Converte uma palavra ou string inteira em uma lista de números correspondentes às posições das letras.
def encode(w):
    return [encodeOne(i) for i in list(w)]

#Faz o inverso de encodeOne: converte um número 0–25 de volta para letra maiúscula.
def decodeOne(n):
    assert 0 <= n < 26
    return chr(ord('A') + n)

#Recebe uma lista de números e retorna a string de letras correspondente.
def decode(w):
    return "".join([decodeOne(n) for n in w])

def cipherMonoOne(c, key):
    assert c in string.ascii_uppercase
    return key[ord(c) - ord('A')]

#Faz substituição monoalfabética de uma letra:
# 1 - Converte c em número de 0–25 (ord(c) - ord('A')).

# 2- Usa este número como índice para pegar a letra correspondente em key.
def cipherMono(w, key):
    return "".join([cipherMonoOne(c, key) for c in w])


#Faz cifra de uma palavra inteira usando substituição monoalfabética.
def back_from_number(lst):
    orda = ord('a')
    foo = [chr(orda+c-1) for c in lst]
    return "".join(foo)

def inverseKey(key):
    nk = [0 for _ in range(26)]
    l = 0
    for i in key:
        nk[encodeOne(i)] = chr(ord('A')+l)
        l += 1
    return "".join(nk)

def stats(text):
    assert len([True for c in text if c in string.ascii_uppercase]) == len(text)
    st = [0 for i in range(26)]
    for c in text:
        st[encodeOne(c)] += 1
    l = sum(st)
    return [x for x in map(lambda x: float(x/l),st)]
```
