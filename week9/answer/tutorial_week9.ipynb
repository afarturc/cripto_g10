### Question 3

For this situation, if what is shared between Alice and Bob is $x^{g} \bmod p$, that is, Bob sends $B = b^{g} \bmod p$ and Alice sends $A = a^{g} \bmod p$, there is a way to agree on a key.

Since $A = a^{g} \bmod p$, and $B = b^{g} \bmod p$, we could:

$A \cdot B = a^{g} \cdot b^{g} = (a \cdot b)^{g} \pmod{p}$


But there is a HUGE problem around here. Alice, as a passive attacker, can break this system very easily by just simply calculating $A \cdot B$ since these are both public values. She does not even need to find the secret numbers $a$ and $b$.

Nevertheless, Alice can still calculate the secret numbers if she wanted to, she needs to compute still a discrete logarith, but that is the **reverse** of the standard Diffie-Hellman problem presented before.

She needs to find $x$ such that $x^{g} \equiv A \pmod{p}$.

Let's implement an snippet that sniff both $a$ and $b$

```python
from sage.all import Mod, gcd, xgcd, Integer

# Let's use some example values for this one
p_ex = 23
g_ex = 5
a_secret = 7
b_secret = 11

# Alice sends A = a^g mod p
A_public = power_mod(a_secret, g_ex, p_ex)

# Alice receives B = a^g mod p
B_public = power_mod(b_secret, g_ex, p_ex)

print(f"Public value A = {a_secret}^{g_ex} mod {p_ex} = {A_public}")
print(f"Public value B = {b_secret}^{g_ex} mod {p_ex} = {B_public}")

# Let's use Sage's built-in method nth_root

R_a = Mod(A_public, p_ex)
R_b = Mod(B_public, p_ex)

try:
    a_found = Integer(R_a.nth_root(g_ex))
    b_found = Integer(R_b.nth_root(g_ex))
    print(f"Eve finds: a = {a_found}, b = {b_found}")
    print(f"Verification: {a_found}^{g_ex} mod {p_ex} = {power_mod(a_found, g_ex, p_ex)}")
    print(f"{b_found}^{g_ex} mod {p_ex} = {power_mod(b_found, g_ex, p_ex)}")
except ValueError as e:
    print(f"No solution found: {e}")
```
