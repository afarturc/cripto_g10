### Question 1

#### a. Describe what is the accepted procedure of a client when receives a public-key certificate


We need to follow the public key infrastructure (PKI), it is a set of roles, policies, hardware, software and procedures needed to create, manage, distribute, use, store and revoke digital certificates.

First of all, all public keys must be encoded in X.509 certificates. Some of these contain the public keys of the Certificate Authorities (CA).

With these CA's public keys, and agent like Bob can verify the signature on Alice's (another agent) certificate. If the certificate is correct, the Bob can use Alice's public key.

Up to this point we assume that, first, Bob has the certificate for CA that issued Alice's certificate and, second, that Bob trusts CA to have checked data on Alice's certificate.

In terms of how does Bob know to trust or not CA, it is reduced to the simplest, he gets the certificate directly from CA and it is assumed that he completely trusts CA's certificate.


#### b. Describe the process of certificate revocation and what are the possible reasons to apply it.

This is formally done by using the Certificate Revokation List (CRL).

The CA periodically publishes a black-list of revoked certificates, the consumers **must** check most-recent CRL periodically. Exceptional CRL may also be published, as best-effort.

The revokation information is usually got from certificate extensions that typically indicate URLs for CRLs.

In the real-world, there is a Trusted Service Provider Lists (TSL), an up-to-date white list of trusted certificates. 

Also, there is an Online Certificate Status Protocol (OCSP). It is a trusted server that checks CRLs for the clients that is usually managed by CAs themselves.

### Question 2

We need to compute the following in order to recover the original message

$ĉ = (c^{Φ(n)}\mod n^2 ),$

$\hat{m}= (ĉ − 1)/n,$

$m = (m̂ Φ(n)^{−1} \mod n).$

```python
from sage.all import *

n = 620496404349687915307910174617

p = factor(n)[0][0]
q = factor(n)[1][0]

print(p, q)

c = 358624662650643040547102063483144791182626860568435345308004

c1 = power_mod(c, (p - 1) * (q - 1), n*n)
print("ĉ =" , c1)

m1 = (c1 - 1) // n
print("m_hat =", m1)


phi_inverse = inverse_mod( (p - 1) * (q - 1), n)
print("phi_inverse =", phi_inverse)

m = m1 * phi_inverse % n

print("m (mensagem) =", m)
```

We can then conclude that it is possible to recover the original message.
